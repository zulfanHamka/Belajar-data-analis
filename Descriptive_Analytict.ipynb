## Rangkuman Penuh: Analisis Deskriptif (*Descriptive Analytics*)

Analisis Deskriptif adalah tahap pertama dan paling mendasar dalam ilmu analitik, yang fokus pada pemahaman apa yang telah terjadi di masa lalu.

### I. Inti Analisis Deskriptif

* **Pertanyaan Kunci:** **"Apa yang telah terjadi?"**
* **Tujuan:** Menjelaskan variansi (penyimpangan) kinerja dan mengidentifikasi tren dalam data historis.
* **Manfaat:** Memungkinkan manajemen untuk mengevaluasi kinerja bisnis perusahaan dan mengambil tindakan korektif yang diperlukan.

### II. Dua Teknik Utama

Analisis deskriptif umumnya diimplementasikan melalui dua metode utama:

#### 1. Analisis Historis (*Historical Analysis*)

* **Fokus:** Menganalisis data masa lalu (internal maupun eksternal) dalam jangka panjang untuk mengungkap pendorong kinerja bisnis (*key drivers*).
* **Contoh Visual (Evolusi Profitabilitas):** Analisis menunjukkan bahwa meskipun pendapatan tumbuh stabil, fluktuasi biaya telah menyebabkan tren penurunan Margin Bersih (*Net Margin*) dari 33% menjadi 27% sejak tahun 2016, Gambar 1 di atas.
* **Kelebihan:** Sederhana, mudah diterima, dapat digunakan untuk *benchmarking*, dan menjadi dasar untuk pembentukan hipotesis.
* **Kekurangan:** Tidak memiliki daya prediksi, karena kinerja masa lalu tidak menjamin masa depan. Data bisa terdistorsi oleh peristiwa luar biasa (misalnya, merger atau event besar).

#### 2. Analisis Variansi (*Variance Analysis*)

* **Fokus:** Membandingkan angka **Aktual** (*Actuals*) dengan **Anggaran/Rencana** (*Budget*) untuk menjelaskan mengapa hasil bisnis berbeda dari harapan.
* **Konsep Favorable/Unfavorable:**
    * **Favorable (Menguntungkan):** Hasil aktual lebih baik dari anggaran (misalnya, pendapatan lebih tinggi, atau biaya lebih rendah).
    * **Unfavorable (Tidak Menguntungkan):** Hasil aktual lebih buruk dari anggaran (misalnya, pendapatan lebih rendah, atau biaya lebih tinggi).

### IV. Keuntungan dan Keterbatasan

* **Keuntungan:** Sangat sederhana, diterima luas, dan penting untuk memicu investigasi ketika terjadi variansi signifikan.
* **Keterbatasan:** Analisis variansi seringkali tidak mengungkapkan **akar penyebab** (*root cause*) masalah, sehingga perlu dikombinasikan dengan pengetahuan bisnis (*business intelligence*) yang mendalam untuk menghasilkan wawasan yang berguna.
